Se um número inválido de código de produto for digitado, o programa deve ignorar, mostrando uma breve mensagem de aviso e não computando o voto, voltando a pedir outro número. Após o final da pesquisa, o programa deverá mostrar:O total de pessoas que votaram nos produtos;A porcentagem de votos de cada um destes produtos;O produto com a menor votação da pesquisa de opinião

```python
def recebeVotos():
  l_votos = []
  sair = 'n'
  for i in range(23):
    l_votos.append(0)
  while sair == 'n':
    voto = int(input('Vote em um produto: '))
    if voto < 0 or voto > 23:
      print('Produto inválido')
    else:
      voto = voto - 1
      l_votos[voto] = l_votos[voto] + 1
    sair = input('Deseja sair? ')
  return l_votos 

def mostraVotos(l_votos):
  soma = 0
  for i in l_votos:
    soma = soma + i
  print('Total de votos: ', soma)
  for i in range(len(l_votos)):
    print('% do produto', i+1,' - ', (l_votos[i]/soma)*100,'%' )

def mostraMenor(l_votos):
  menor = 0
  for i in range(len(l_votos)):
    if l_votos[i] < l_votos[menor]:
      menor = i
  print('Produto menos votado foi o ', menor+1)

def principal():
  lista_votos = recebeVotos()
  mostraVotos(lista_votos)
  mostraMenor(lista_votos)
  
principal()
```

```python
def Jaccard(a, b):
  a = set(a)
  b = set(b)
  return len(a.intersection(b))/len(a.union(b))

def calculaMatriz(user, mat):
  l_dist = []
  for m in mat:
    l_dist.append(Jaccard(user, m))
  return l_dist

def maiorMatriz(lista):
  maior = 0
  for i in range(len(lista)):
    if lista[i] > lista[maior]:
      maior = i
  return maior

def principal():
  secoes = ['bom','ruim','neutro']
  matriz = [ 
            ['feliz','alegre','dinheiro','sorrir'],
            ['estressado','saco','morreu','devendo'],
            ['guloseimas','festa','criança','sobremesa']
  ]
  usuario = input('Tweet: ').lower().split(' ')
  calculo = calculaMatriz(usuario, matriz)
  print(calculo)
  maior = maiorMatriz(calculo)
  print(secoes[maior])

principal()
```

•Faça um algoritmo que, dada uma
matriz de 20x20 com valores inteiros preenchida, determine:


•A) O maior valor da matriz


•B) A média dos valores


•C) A soma total de cada linha da
matriz


•Obs.:  deverá ser utilizada função.

```python
import random

def geraMatriz():
  matriz = [[0]*20 for i in range(20)]
  for l in range(len(matriz)):
    for c in range(len(matriz[l])):
      matriz[l][c] = random.randint(1, 10)
  return matriz

def maior(matriz):
  maior = matriz[0][0]
  for i in matriz:
    for valor in i:
      if valor > maior:
        maior = valor
  return maior

def media(matriz):
  soma = 0
  for l in range(len(matriz)):
    for c in range(len(matriz[l])):
      soma = soma + matriz[l][c]
  media = soma/(len(matriz) * len(matriz[0]))
  return media

def soma(matriz):
  l_linha = []
  for l in range(len(matriz)):
    soma = 0
    for c in range(len(matriz[l])):
      soma = soma + matriz[l][c]
    l_linha.append(soma)
  return l_linha

def main():
  m = geraMatriz()
  ma = maior(m)
  me = media(m)
  s = soma(m)
  print(m)
  print(f'O maior valor é o {ma}')
  print(f'A média dos valores é {me}')
  print(f'A soma das linhas é {s}')

main()
```
